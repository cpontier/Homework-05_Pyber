Your objective is to build a [Bubble Plot](https://en.wikipedia.org/wiki/Bubble_chart) that showcases the relationship between four key variables:

* Average Fare ($) Per City
* Total Number of Rides Per City
* Total Number of Drivers Per City
* City Type (Urban, Suburban, Rural)


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from scipy import stats
import seaborn as sns
sns.set()
```

```python
citydata = os.path.join('raw_data', 'city_data.csv')
ridedata = os.path.join('raw_data', 'ride_data.csv')
```

```python
citydf = pd.read_csv(citydata)
ridedf = pd.read_csv(ridedata)
citydf.head(5)
```

```python
ridedf.head(5)
```

```python
#extracting necessary information from the given files

averagefare = ridedf.groupby(['city'])['fare'].mean()
totalrides = ridedf.groupby(['city'])['ride_id'].count()
totaldrivers = citydf[['city','driver_count']]
citytype = citydf.groupby(['type'])['type'].count()
citydrive = citydf.groupby(['type'])['driver_count'].sum()
```

```python
#merge dfs for getting city type into the df
totaldf = pd.merge(ridedf, citydf, on='city')

totaldf.head()

```

Your objective is to build a Bubble Plot that showcases the relationship between four key variables:

Average Fare ($) Per City
Total Number of Rides Per City
Total Number of Drivers Per City
City Type (Urban, Suburban, Rural)

```python
# create data

averagefaredf = list(totaldf.groupby(['city'])['fare'].mean())
totalridedf = list(totaldf.groupby(['city'])['ride_id'].count())
totaldriverdf = list(totaldf.groupby(['city'])['driver_count'].max())
citytypedf = list(totaldf.groupby(['city'])['fare'].sum())

# generate list of unique cities

citylistx = set(list(totaldf['city']))

# set x, labels and limit to exclude outliers

x = totalridedf
plt.xlabel('Number of Rides', fontsize = 20)
plt.xlim(0, np.mean(totalridedf)+3*np.std(totalridedf))

#set y, allow outliers in axis, 

y = averagefaredf
plt.ylabel('Average Fare', fontsize = 20)
plt.ylim(0, max(averagefaredf)+2)

#set z, which will be used to determine size of points

z = citytypedf

#adding text to annotate the information presented
 
# use the scatter function
plt.scatter(x, y, s=z*5000, alpha=0.7, c=totaldriverdf, cmap="Blues")
plt.rcParams["figure.figsize"] = [16,9]

# try to add a colorbar
plt.colorbar(label='Drivers in City', fontsize = 15)

# adding a legend
for area in [round(np.min(citytypedf)), round(np.mean(citytypedf))]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
                label=str(area) + ' Drivers')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Total Fares', fontsize=15)

#adding a title
plt.title('Ride Metrics by City Type', fontsize = 25)

plt.show()


```

In addition, you will be expected to produce the following three pie charts:

* % of Total Fares by City Type
* % of Total Rides by City Type
* % of Total Drivers by City Type


```python
#find sum of fares per type
typefaresdf = totaldf.groupby(['type'])['fare'].sum()
```

```python
cityfarepiece = typefaresdf.plot(kind='pie', 
                        startangle=90, 
                        colormap="Blues", 
                        explode=(.2, 0, 0),
                        autopct='%1.1f%%',
                        fontsize = 20,
                        labels = ('','',''))
plt.title('City Type v Total Fares', fontsize = 20)
cityfarepiece.set_ylabel('')
plt.legend(fontsize = 20, labels = ("Rural", "Suburban", "Urban"))
plt.axis('equal')
plt.show()
```

```python
citypie = citytype.plot(kind='pie', 
                        startangle=90, 
                        colormap="Blues", 
                        explode=(.2, 0, 0),
                        autopct='%1.1f%%',
                        fontsize = 20,
                        labels = ('','',''))
citypie.set_ylabel('')
plt.legend(fontsize = 20, labels = ("Rural", "Suburban", "Urban"))
plt.title('City v Rides', fontsize = 20)
plt.axis('equal')
plt.show()
```

```python
citydriverpie = citydrive.plot(kind='pie', 
                        startangle=90, 
                        colormap="Blues", 
                        explode=(.2, 0, 0),
                        autopct='%1.1f%%',
                        fontsize = 20,
                        labels = ('','',''))
citydriverpie.set_ylabel('')
plt.legend(fontsize = 20, labels = ("Rural", "Suburban", "Urban"))
plt.title('City Type v Driver Count')
plt.axis('equal')
plt.show()
```
